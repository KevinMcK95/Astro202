I will create a nice plot below:

```python
x = np.arange(-3*np.pi,3*np.pi,0.1)
y = 2*np.sin(x)+x/5

plt.figure()
plt.plot(x,y,label='Line 1')
plt.plot(x,-2*y,label='Line 2')
plt.xlabel('x Values')
plt.ylabel('y Values')
plt.title('Simple Plot Example')

leg = plt.legend(loc='best')
leg.get_frame().set_alpha(0.5)

plt.show()
```
